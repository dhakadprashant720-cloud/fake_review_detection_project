```python
import numpy as np

class MyLogisticRegression:
    def __init__(self, lr=0.01, iterations=1000):
        self.lr = lr
        self.iterations = iterations
        self.W = None
        self.b = None

    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def predict(self, X):
        linear = X.dot(self.W) + self.b
        y_pred = self._sigmoid(linear)
        return (y_pred >= 0.5).astype(int), y_pred
```

```python
import pickle

# load model
with open("logistic_model.pkl", "rb") as f:
    model = pickle.load(f)

# load vectorizer
with open("tfidf_vec.pkl", "rb") as f:
    vectorizer = pickle.load(f)
```

```python
print(type(model))
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab') # Added to resolve LookupError

```

```python
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = word_tokenize(text)
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
#
# input
review = input("Enter your review: ")

# preprocess
clean_review = preprocess_text(review)

# tfidf
review_vector = vectorizer.transform([clean_review])

# predict
pred, prob = model.predict(review_vector)

# output
if pred[0] == 1:
    print(f" Fake Review ({prob[0]*100:.2f}% confidence)")
else:
    print(f" Real Review ({(1-prob[0])*100:.2f}% confidence)")
```
